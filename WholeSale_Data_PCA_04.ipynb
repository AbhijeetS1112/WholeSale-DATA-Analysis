```python
# Splitting The Data into Dependent variable and Independent variable

x=df1.drop(["Channel"],axis=1)
y=df1["Channel"]
```

```python
# Split The data Model into training And Test
# For Building The Model 
# And Evaluation The Model

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.75,random_state=111)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# First Classification Model : Logit
from sklearn.linear_model import LogisticRegression
logit=LogisticRegression()
logit.fit(x_train,y_train)
```

```python
y_pred_logit_train=logit.predict(x_train)
y_pred_logit_test=logit.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print(classification_report(y_train,y_pred_logit_train))
print("####"*25)
print(classification_report(y_test,y_pred_logit_test))
```

```python
print(confusion_matrix(y_train,y_pred_logit_train))
print("####"*25)
print(confusion_matrix(y_test,y_pred_logit_test))
```

```python
print(accuracy_score(y_train,y_pred_logit_train))
print("####"*25)
print(accuracy_score(y_test,y_pred_logit_test))
```
